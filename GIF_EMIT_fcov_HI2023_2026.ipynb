```python
import glob
import os
import numpy as np
from PIL import Image
import imageio.v2 as imageio

# --- CONFIG ---
input_folder = "/Users/gillopez/Downloads/testing1/Images"   # folder with your copied images
file_pattern = "*.png"
output_gif = "output.gif"
fps = 1 # frames per second
loop_count = 0  # 0 = infinite loop


# --- GET FILES IN ORDER ---
image_files = sorted(glob.glob(f"{input_folder}/{file_pattern}"))
print(f"Found {len(image_files)} Images")

# --- CHECK SIZES (optional, just to see what you're working with) ---
for f in image_files:
    with Image.open(f) as img:
        print(os.path.basename(f), img.size)

# --- PICK A TARGET SIZE ---
# Option A: use the size of the first image
with Image.open(image_files[0]) as first_img:
    target_size = first_img.size  # (width, height)

# Option B: hardcode a size instead, e.g.:
# target_size = (800, 600)

# --- RESIZE + LOAD ---
frames = []
for f in image_files:
    img = Image.open(f).convert("RGB").resize(target_size, Image.LANCZOS)
    frames.append(np.array(img))

# --- SAVE AS LOOPING GIF ---
imageio.mimsave(output_gif, frames, fps=fps, loop=loop_count)
print(f"GIF saved to {output_gif}")
```
